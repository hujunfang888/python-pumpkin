### Question 6
Using the startwars dataset imported below, identify the individuals with tallest height and lowest mass.

````python
import pandas as pd
starwars_dataset="https://www.fabricionarcizo.com/post/starwars/updated_starwars.csv"
starwars = pd.read_csv(starwars_dataset)
starwars['height'] = pd.to_numeric(starwars['height'], errors='coerce')
starwars['mass'] = pd.to_numeric(starwars['mass'], errors='coerce')
tallest_individual = starwars.loc[starwars['height'].idxmax()]
lowest_mass_individual = starwars.loc[starwars['mass'].idxmin()]
print("Individual with the tallest height:")
print(tallest_individual)

print("\nIndividual with the lowest mass:")
print(lowest_mass_individual)
````

### Question 8
Using the diamonds dataset and the plotting code of your choice, plot the following from the diamonds dataset.

1. Carat vs Price in a scatter plot.
2. A histogram of prices.
3. A boxplot of cut vs price.

The code can be loaded from seaborn in the cell below.

````python
diamonds = sns.load_dataset('diamonds')
````

````python
import seaborn as sns
import matplotlib.pyplot as plt

diamonds = sns.load_dataset('diamonds')


plt.figure(figsize=(9, 6))  
sns.scatterplot(x='carat', y='price', data=diamonds, alpha=0.7)
plt.title('Scatter Plot: Carat vs Price', fontsize=12)
plt.xlabel('Carat', fontsize=12)
plt.ylabel('Price', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()


plt.figure(figsize=(9, 6))  
sns.histplot(diamonds['price'], bins=50, kde=True, color='blue')
plt.title('Histogram of Prices', fontsize=12)
plt.xlabel('Price', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()


plt.figure(figsize=(9, 6)) 
sns.boxplot(x='cut', y='price', data=diamonds, palette='Set2')
plt.title('Boxplot: Cut vs Price', fontsize=12)
plt.xlabel('Cut', fontsize=12)
plt.ylabel('Price', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
````

### Question 10
Write a function that will convert carat to grams and find the heaviest diamond.

1 carat = 0.2 grams

````python
import seaborn as sns
diamonds = sns.load_dataset('diamonds')
def find_heaviest_diamond(data):
    data['weight_in_grams'] = data['carat'] * 0.2
    heaviest_diamond = data.loc[data['weight_in_grams'].idxmax()]
    return heaviest_diamond
heaviest_diamond = find_heaviest_diamond(diamonds)
print("The heaviest diamond is:")
print(heaviest_diamond)
````
